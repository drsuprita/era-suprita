```python
def mean(array):
    n=len(array)
    return np.sum(array)/n
mean(np.array([1,2,3,4,5]))
```

```python
def variance(array):
    n=len(array)
    m=mean(array)
    return np.sum((array-m)**2)/n  
variance(np.array([1,2,3,4,5]))  
```
